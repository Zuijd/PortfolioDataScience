##### High correlation - not expected
According to the schema the minutes played per yellow card have a correlation of 0.74 with the amount of clean sheets. The fitting scatter plot can be found below. The outcome of this correlation is surprising, there can be found a good corelation between the two columns. There can be seen that, for some of the players, the higher the minutes played per yellow card the more clean sheet the player gets. This statement only stands for a select group of players, which is logical because some players get lots of yellow cards and some players get none in a game. These players can be on the same team and so, if a whole team gets a clean sheet the chance is small that all players of that team received a yellow card that game. Therefore the highly populated bottom right of the plot can be explained.    

```python
playerStatsAllLeagues.plot(kind='scatter', x='Yellow Minutes Played', y='CS')
```
