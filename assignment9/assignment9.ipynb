```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
```

### Correlation:

```python
penguins.corr().style.background_gradient(cmap='flare', axis=None).set_precision(2)
```

Dit resultaat is naar verwachting, aangezien het te verwachten is dat de lengte van de flipper groeit wanneer de algemene lichaamsmassa groter wordt. Verder is het wel vreemd dat de lengte van de bek in slechte correlatie is ten opzichte van de diepte van de bek.

### Scatter plots
#### Strongest positive correlation: 

```python
penguins.plot(kind='scatter', x='body_mass_g', y='flipper_length_mm')
```

#### Strongest negative correlation: 

```python
penguins.plot(kind='scatter', x='bill_depth_mm', y='flipper_length_mm')
```

```python
penguins.plot(kind='scatter', x='bill_depth_mm', y='bill_length_mm')
```
