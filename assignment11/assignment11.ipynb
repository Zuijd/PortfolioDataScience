```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.groupby('island').mean()
```

#### barplots
##### bill_length_mm


```python
sns.barplot(y="bill_length_mm", x="island", data=penguins)
```

In the barplot is visible that the confidence interval for the island 'Torgersen', is not overlapping with the other islands. So it can be said that there is a statistically significant difference between 'Torgersen' and the other islands. The intervals og the other two islands: 'Biscoe' and 'Dream' are overlapping, so there is no statistically significant difference between these two islands. 

##### bill_depth_mm

```python
sns.barplot(y="bill_depth_mm", x="island", data=penguins)
```

The confidence intervals of the islands: 'Torgersen' and 'Dream' are overlapping eachother, so it can be said that there is not a statistically significant difference between these island. But between 'Biscoe' and the other two islands there can be found a statistically significant difference, because the intervals are nowhere near eachother.
